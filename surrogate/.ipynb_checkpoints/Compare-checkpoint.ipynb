```python
with open('train_eval_acc_ste.npy', 'rb') as f:
    train_ste = np.load(f)
    eval_ste = np.load(f)
    acc_ste = np.load(f)
```

```python
plt.plot(np.log(train_exact[0:1000]), color='C0',alpha=0.8)
plt.plot(np.log(train_fast[0:1000]), color='C1', alpha=0.8)
plt.plot(np.log(train_sre[0:1000]), color='C3', alpha=0.8)
plt.plot(np.log(train_sigmoid[0:1000]), color='gray', alpha=0.2)
plt.plot(np.log(train_ste[0:1000]), color='gray', alpha=0.2)

plt.rcParams.update({'font.size': 14})

plt.legend(['exact', 'fast sigmoid', 'SRE', 'sigmoid', 'STE'])
plt.xlabel('Training step / 10')
plt.ylabel('Cross Entropy Loss (in Log Scale)')
plt.title('Training Loss Over Number of Steps')
plt.savefig('comparing_surrogate.png', dpi=200)
```
