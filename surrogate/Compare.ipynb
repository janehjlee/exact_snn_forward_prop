```python
N=100
```

```python
plt.figure(figsize=(10,8))

plt.plot((train_exact), color='C0', alpha=0.1, label='_nolegend_')
plt.plot(uniform_filter1d(train_exact, N), color='C0')
plt.plot((train_fast_count), color='C1', alpha=0.1, label='_nolegend_')
plt.plot(uniform_filter1d(train_fast_count, N), color='C1')
plt.plot((train_fast_rate[0]), color='C2', alpha=0.1, label='_nolegend_')
plt.plot(uniform_filter1d(train_fast_rate[0], N), color='C2')
plt.plot((train_ste_count), color='C3', alpha=0.1, label='_nolegend_')
plt.plot(uniform_filter1d(train_ste_count, N), color='C3')
plt.plot((train_ste_rate[0]), color='C4', alpha=0.1, label='_nolegend_')
plt.plot(uniform_filter1d(train_ste_rate[0], N), color='C4')

plt.rcParams.update({'font.size': 20})

plt.legend(['exact', 'fast count', 'fast rate', 'STE count', 'STE rate'])
plt.xlabel('Training steps', fontsize=22)
plt.ylabel('Cross Entropy Loss', fontsize=22)
plt.title('Training Loss Over Number of Steps')
plt.savefig('comparing_surrogate.png', dpi=200)
```
